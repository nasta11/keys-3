```python
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Загрузка данных
ticker = 'GAZP.ME'
data = yf.download(ticker, start="2018-01-01", end="2023-01-01")

# Проверяем, что данные были успешно загружены
print(data.tail())

# Выборка данных о цене закрытия
close_prices = data['Close'].values.reshape(-1, 1)

# Масштабирование данных
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(close_prices)

# Функция для создания временных последовательностей
def create_sequences(data, sequence_length):
    xs = []
    ys = []
    for i in range(len(data) - sequence_length):
        sequence = data[i:(i + sequence_length)]
        label = data[i + sequence_length]
        xs.append(sequence)
        ys.append(label)
    return np.array(xs), np.array(ys)

# Длина временной последовательности
sequence_length = 60

# Создание временных последовательностей
X, y = create_sequences(scaled_data, sequence_length)

# Разделение данных на обучающую и тестовую выборки
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Размеры входных данных
input_shape = (X_train.shape[1], X_train.shape[2])

# Создание модели
model = Sequential([
    LSTM(units=50, return_sequences=True, input_shape=input_shape),
    Dropout(0.2),
    LSTM(units=50, return_sequences=False),
    Dropout(0.2),
    Dense(units=25),
    Dense(units=1)
])

# Компиляция модели
model.compile(optimizer='adam', loss='mean_squared_error')

# Обучение модели
model.fit(X_train, y_train, batch_size=32, epochs=100, validation_split=0.1)
```

```python
import matplotlib.pyplot as plt

history = model.fit(X_train, y_train, batch_size=32, epochs=100, validation_data=(X_test, y_test))

# Визуализация истории обучения
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Обучение - потери')
plt.plot(history.history['val_loss'], label='Валидация - потери')
plt.title('История обучения модели')
plt.ylabel('Потери')
plt.xlabel('Эпоха')
plt.legend()
plt.show()
```

```python
predictions = model.predict(X_test)
```

```python
# Обратное масштабирование предсказаний
predicted_prices = scaler.inverse_transform(predictions)

# Обратное масштабирование фактических значений
actual_prices = scaler.inverse_transform(y_test.reshape(-1, 1))
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

rmse = sqrt(mean_squared_error(actual_prices, predicted_prices))
print(f"RMSE: {rmse}")
```

```python
import matplotlib.pyplot as plt

# Визуализация фактических и предсказанных цен
plt.figure(figsize=(10, 6))
plt.plot(actual_prices, color='blue', label='Фактические цены')
plt.plot(predicted_prices, color='red', linestyle='--', label='Предсказанные цены')
plt.title('Сравнение фактических и предсказанных цен')
plt.xlabel('Время')
plt.ylabel('Цена')
plt.legend()
plt.show()
```
